### Example 18.1 Linear Interpolation

#### Problem Statement : 
> - Estimate the natural logarithm of 2 using linear interpolation. 
First, perform the computation by interpolating between ln 1 5 0 and ln 6 5 1.791759. 
> - Then, repeat the procedure, but use a smaller interval from ln 1 to ln 4 (1.386294). Note 
that the true value of ln 2 is 0.6931472.

```python
Latex(f""" \\begin{{equation*}} \\huge
      f(x) = \\ln(x)
      \\end{{equation*}} 
      """)
```

```python
Latex(""" \\begin{equation*} \\huge
      f_1(x) = f(x_0) + f'(x_0)(x-x_0) = f(x_0) + {f(x_1) - f(x_0)\\over x_1 - x_0 }(x-x_0)
      \\end{equation*} 
      """)
```

```python
def func(x):
    return np.log(x)
```

```python
xi = np.array([i*0.1 for i in range(1, 100)])
yi = np.array([func(i) for i in xi])

fig, ax = plt.subplots()
line, = ax.plot(xi, yi, lw=2)
ax.set_ylim(0, 2)
```

```python
df = func(6) - func(1)
dx = 6-1
a = df/dx
print(f'df = {df}, dx = {dx}, a = {a}')

x0,y0 = 2,func(2)
x1,y1 = 2, a*(2-1)
```

```python
fig, ax = plt.subplots()
line, = ax.plot(xi, yi, lw=2)
line2, = ax.plot(xi, a*(xi-1), lw=2)
ball = plt.Circle((x0, y0), 0.10,color='r')
ball2 = plt.Circle((x1, y1), 0.10,color='b')
ax.add_patch(ball)
ax.add_patch(ball2)
ax.set_ylim(0, 2)

```

```python
df = func(4) - func(1)
dx = 4-1
a = df/dx
print(f'df = {df}, dx = {dx}, a = {a}')

x2,y2 = 2, a*(2-1)
```

```python
fig, ax = plt.subplots()
line, = ax.plot(xi, yi, lw=2)
line2, = ax.plot(xi, a*(xi-1), lw=2)
ball = plt.Circle((x0, y0), 0.10, color='r')
ball2 = plt.Circle((x1, y1), 0.10, color='b')
ball3 = plt.Circle((x2, y2), 0.10, color='g')
ax.add_patch(ball)
ax.add_patch(ball2)
ax.add_patch(ball3)
ax.set_ylim(0, 2)

```

### Example 18.2 Quadratic Interpolation

#### Problem Statement : 
> - Fit a second-order polynomial to the three points used in Example 18.1: 
>   - x0 = 1 f(x0) = 0
>   - x1 = 4 f(x1) = 1.386294
>   - x2 = 6 f(x2) = 1.791759
> - Use the polynomial to evaluate ln 2.

```python
Latex(f""" \\begin{{equation*}} \\huge
      f(x) = \\ln(x)
      \\end{{equation*}} 
      """)
```

```python
Latex(f""" \\begin{{equation*}} \\huge
      f_2(x) = b_0 + b_1(x-x_0)+b_2(x-x_0)(x-x_1)
      \\end{{equation*}} 
      """)
```

```python
Latex(f""" \\begin{{equation*}} \\huge
        f_2(x_0) = b_0 + b_1(x_0-x_0)+b_2(x_0-x_0)(x_0-x_1)=0
        \\end{{equation*}} 
        """)
```

```python

Latex(f""" \\begin{{equation*}} \\huge
         b_0 =0
        \\end{{equation*}} 
        """)
```

```python
Latex(f""" \\begin{{equation*}} \\huge
        f_2(x_1) = b_1(x_1-x_0)+b_2(x_1-x_0)(x_1-x_1)=1.386294
        \\end{{equation*}} 
        """)
```

```python

Latex(""" \\begin{equation*} \\huge
         b_1 = {1.386294 \\over 4-1}=0.462
        \\end{equation*} 
        """)
```

```python
Latex(f""" \\begin{{equation*}} \\huge
        f_2(x_2) = b_1(x_2-x_0)+b_2(x_2-x_0)(x_2-x_1)=1.791759
        \\end{{equation*}} 
        """)
```

```python

Latex(""" \\begin{equation*} \\huge
         b_2 = {{1.791756-1.386294 \\over 6-4} - 0.462 \\over 6-1}=-0.0518731
        \\end{equation*} 
        """)
```

```python
def func(x):
    return np.log(x)

b = np.array([0,0.462,-0.0518731])

def func2(x):
    return b[1]*(x-1)+b[2]*(x-1)*(x-4)
```

```python
xi = np.array([i*0.1 for i in range(1, 100)])
yi = np.array([func(i) for i in xi])
yi2 = np.array([func2(i) for i in xi])

fig, ax = plt.subplots()
line, = ax.plot(xi, yi, lw=2,color='b')
line2, = ax.plot(xi, yi2, lw=2,color='r')
line3, = ax.plot(xi, a*(xi-1), lw=2,color='g')
ball = plt.Circle((2, func(2)), 0.10, color='b')
ball2 = plt.Circle((2, func2(2)), 0.10, color='r')
ball3 = plt.Circle((2, a*(2-1)), 0.10, color='g')
ax.add_patch(ball3)
ax.add_patch(ball2)
ax.add_patch(ball)

ax.set_ylim(0, 3)
```

### Example 18.6 Quadratic Interpolation

#### Problem Statement : 
> - Use a Lagrange interpolating polynomial of the first and second 
order to evaluate ln 2 on the basis of the data given in Example 18.2:: 
>   - x0 = 1 f(x0) = 0
>   - x1 = 4 f(x1) = 1.386294
>   - x2 = 6 f(x2) = 1.791759


```python
Latex(""" \\begin{equation*} \\huge
         f_n(x) = \\sum_{i=0}^{n} L_i(x)f(x_i), L_i(x) = \\prod_{j=0,j\\neq i}^{n} {x-x_j \\over x_i-x_j}
        \\end{equation*} 
        """)
```

```python
def L_n(x,xi,i,n):
    assert n <= len(xi)
    L = 1
    for j in range(n):
        if j != i:
            L *= (x-xi[j])/(xi[i]-xi[j])
    return L

def f_n(x,xi,yi,n):
    
    L = np.array([L_n(x,xi,i,n+1) for i in range(n+1)])
    return np.dot(L,yi[:n+1])
```

```python
xi = np.array([i*0.1 for i in range(1, 100)])
yi = np.array([func(i) for i in xi])

fig, ax = plt.subplots()
line, = ax.plot(xi, yi, lw=2,color='b')
```

```python
xp = np.array([1,4,6])
yp = np.array([func(x) for x in xp])
```

```python
f_1_2  = f_n(2,xp,yp,1)
f_2_2  = f_n(2,xp,yp,2)
print(f_1_2,f_2_2)

xi = np.array([i*0.1 for i in range(1, 100)])
yi = np.array([func(i) for i in xi])

fig, ax = plt.subplots()
line, = ax.plot(xi, yi, lw=2,color='b')
plt.ylim(0,3)
```

```python
fig, ax = plt.subplots()
line, = ax.plot(xi, yi, lw=2,color='b')
line2, = ax.plot(xi, [f_n(x,xp,yp,1) for x in xi], lw=2,color='g')
line3, = ax.plot(xi, [f_n(x,xp,yp,2) for x in xi], lw=2,color='r')

ball1 = plt.Circle((2, f_1_2), 0.10, color='g')
ball2 = plt.Circle((2, f_2_2), 0.10, color='r')

ax.add_patch(ball1)
ax.add_patch(ball2)
plt.ylim(0,3)
```
